# Using histogram to represent relationship between all features

```python
data.hist(figsize=(50,30))
plt.show()
```

# Countries affected by terrorist attacks

```python
data['country'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=data['country_txt'].value_counts()[:20].index, y=data['country_txt'].value_counts()[:20].values)
plt.title('Top countries got attacked')
plt.xticks(rotation = 70)
plt.xlabel('Countries')
plt.ylabel('Number of attacks')
plt.show()
```

# Region affected by terrorist attacks

```python
data['region_txt'].value_counts()
plt.figure(figsize=(20,10))
sns.countplot(x='region_txt', data= data)
plt.xticks(rotation= 70)
plt.show()
```

```python
data['city'].value_counts()
```

# Number of attacks 1970 to 2017

```python
plt.subplots(figsize=(20, 10))
sns.countplot(x='iyear', data=data,palette='tab10')
plt.xticks(rotation=70)
plt.xlabel('Terror attacks from 1970 to 2017')
plt.show()
```

# Type of attack used by terrorists

```python
plt.figure(figsize=(10,10))
sns.countplot(x=data['attacktype1_txt'], order=data['attacktype1_txt'].value_counts().index, palette='rainbow')
plt.xticks(rotation=90)
plt.xlabel('Attack type')
plt.ylabel('Number of attack')
plt.show()
```

# Attack Target

```python
plt.figure(figsize=(20,10))
sns.countplot(x=data['targtype1_txt'], order=data['targtype1_txt'].value_counts().index, palette='flare')
plt.xticks(rotation =90)
plt.xlabel('Target')
plt.ylabel('Number of target')
plt.show()
```

# Name of group attacks terrorist

```python
plt.figure(figsize=(15,6))
sns.barplot(x=data['gname'].value_counts()[:20].index, y=data['gname'].value_counts()[:20].values, palette='pastel')
plt.xticks(rotation=90)
plt.title('TOTAL GROUP ATTACKS')
plt.xlabel('GROUP NAME')
plt.ylabel('TOTAL NUMBER OF ATTACKS')
plt.show()
```

# Cities Affected By Terrorist

```python
data['city'].value_counts()
plt.figure(figsize=(20, 10))
sns.barplot(x=data['city'].value_counts()[:20].index, y=data['city'].value_counts()[:20].values)
plt.title('TOP CITIES GOT ATTACKED')
plt.xlabel('CITIES')
plt.ylabel('NUMBER OF ATTACKS')
plt.xticks(rotation=70)
plt.show()
```

# Weapon Used For Supporting The Attack Type Visual

```python
data['weaptype1_txt'].value_counts()
plt.figure(figsize=(20, 10))
color=['lightblue','red','lightpink','brown']
sns.barplot(x=data['weaptype1_txt'].value_counts()[:20].index, y=data['weaptype1_txt'].value_counts()[:20].values)
plt.title('WEAPON USAGE')
plt.xlabel('WEAPON USED')
plt.ylabel('NUMBER OF TIMES')
plt.xticks(rotation=70)
plt.show()
```
